```python
import pandas as pd
```

```python
female_voice = pd.read_csv('Data/male-female-data/male-female-data/FemaleVoice.tsv', sep='\t')
male_voice = pd.read_csv('Data/male-female-data/male-female-data/MaleVoice.tsv', sep='\t')
```

```python
df = pd.concat([female_voice, male_voice], ignore_index=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.to_csv("Data/male-female-data/male-female-data/data.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split df into train and test with test size = 0.2
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_df.to_csv('Data/train.csv',index=False)
test_df.to_csv('Data/test.csv',index=False)
```

```python
import shutil

# Source and destination directories
src_dir = "Data/male-female-data/male-female-data"
dst_dir = "Data/train"

# Create destination directory if it doesn't exist
os.makedirs(dst_dir, exist_ok=True)

# Move all files from source to destination
for filename in train_df['audio_id']:
    print(filename)
    src_path = os.path.join(src_dir, filename+'.wav')
    dst_path = os.path.join(dst_dir, filename+'.wav')
    print(src_path)
    # Check if it is a file (skip directories)
    if os.path.isfile(src_path):
        shutil.move(src_path, dst_path)
        print(f"Moved: {filename}")
```
